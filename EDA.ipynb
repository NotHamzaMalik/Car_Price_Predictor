```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pakwheels_used_car_data_v02.csv")
df = df[['make','year','engine','mileage','transmission','fuel','price']]
df.dropna(inplace=True)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['price'], bins=30)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['year'], y=df['price'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['transmission'], y=df['price'])
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['fuel'], y=df['price'])
plt.show()
```
